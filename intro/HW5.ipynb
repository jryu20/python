## Problem 1: Faceted Histogram

Run the following code block to define a function which generates two 1-dimensional `numpy` arrays. The first array, called `groups`, consists of integers between `0` and `n_groups - 1`, inclusive. The second array, called `data`, consists of real numbers. 

````python
import numpy as np
from matplotlib import pyplot as plt

def create_data(n, n_groups):
    """
    generate a set of fake data with group labels. 
    n data points and group labels are generated. 
    n_groups controls the number of distinct groups. 
    Returns an np.array() of integer group labels and an 
    np.array() of float data. 
    """
    
    # random group assignments as integers between 0 and n_groups-1, inclusive
    groups = np.random.randint(0, n_groups, n)
    
    # function of the groups plus gaussian noise (bell curve)
    data   = np.sin(groups) + np.random.randn(n)
    
    return(groups, data)
````

### Requirements: 

1. Your function should work **whenever `m_rows*m_cols` is equal to the total number of groups.** Your function should first check that this is the case, and raise an informative `ValueError` if not. You may assume that there is at least one data point for each group label in the data supplied. 
2. For full credit, you should not loop over the individual entries of `groups` or `data`. It is acceptable to loop over the distinct values of `groups`. In general, aim to minimize `for`-loops and maximize use of `Numpy` indexing. 
2. Use of `pandas` is acceptable but unnecessary, and is unlikely to make your solution significantly simpler. 
2. You should include a horizontal axis label (of your choice) along **only the bottom row** of axes.

### Hints

- If your plots look "squished," then `plt.tight_layout()` is sometimes helpful. Just call it after constructing your figure, with no arguments. 
- Integer division `i // j` and remainders `i % j` are helpful here, although other solutions are also possible. 


````python
# your solution here
def facet_hist(groups, data, m_rows, m_cols, figsize, **kwargs): #using **kwargs from part b
    """
    plots multiple histograms based on a random number of data points assigned to a random histogram
    """
    for i in range(m_rows*m_cols): #checking if the dimensions match the group size
        if i not in groups: 
            raise ValueError("Your dimensions do not match the size of the groups") #throw an error if mismatch
    
    fig, ax = plt.subplots(m_rows, m_cols, figsize = figsize) #create a figure with the appropriate amount of subplots and figsize
    for i in range(m_rows): #iterate through each row
        for j in range(m_cols): #iterate through each column
            #use boolean indexing below to grab just the data points corresponding to each histogram
            ax[i][j].hist(data[groups==(m_cols*i+j)], **kwargs) #use **kwargs for additional inputs
            ax[i][j].set(title = "Group " + str(m_cols*i+j)) #set a numbered title for each histogram
            if i == m_rows-1: #this determines the bottom row
                ax[i][j].set(xlabel = "Data Points") #set horizontal axis label
            if j == 0: #this determines the leftmost column
                ax[i][j].set(ylabel = "Frequency") #set vertical axis label
````

Write a function called `scatterplot_matrix()`, with arguments `cols` and `figsize`. There is a separate scatterplot for each possible pair of variables. In fact, there are two: one where the first variable is on the horizontal axis, and one where it's on the vertical axis. Some analysts prefer to remove half the plots to avoid redundancy, but you don't have to bother with that. The diagonal is empty, since there's no point in investigating the relationship between a variable and itself. 

Don't forget comments and docstrings! 

````python
# define your function
def scatterplot_matrix(cols, figsize):
    """
    creates a n by n "matrix" of scatterplots that compares one category of the table to another and looks at the correlation
    """
    fig, ax = plt.subplots(len(cols),len(cols),figsize = figsize) #create the figure and subplots (a n by n "matrix")
    for i in range(len(cols)):
        for j in range(len(cols)): #these two for loops ensure that we are hitting every scatterplot
            x = happiness[cols[i]] #set the values to plot in the x-axis
            y = happiness[cols[j]] #set the values to plot in the y-axis
            ax[i][j].set(ylabel = cols[j], title = cols[i]) #label each graph (we have the x-axis label at the top)
            if i == j: #essentially skip over these cases because no point in looking at a plot of a variable against itself
                continue
            my_number = np.corrcoef(x,y)[0,1] #corrcoef function returns a matrix, so we pull just the value that we want
            ax[i][j].set(xlabel = r"$\rho$ = " + str(np.round(my_number, 2))) #label the coefficients on the bottom of each plot
            ax[i][j].scatter(x,y) #graph the plot

cols = ["Score", 
        "GDP per capita", 
        "Social support"]

scatterplot_matrix(cols,figsize = (10,10))
plt.tight_layout()
````

Run the following code to download two time series data sets:

````python
# run this block
# if you experience ConnectionRefused errors, you may instead 
# copy the urls into your browser, save the files as DJI.csv 
# and COVID.csv respectively in the same directory as the notebook.
# Then, in the lines using the function pd.read_csv(), replace 
# the url with "DJI.csv" and "COVID.csv" 

import pandas as pd
import datetime

url = "https://query1.finance.yahoo.com/v7/finance/download/%5EDJI?period1=1580750232&period2=1712372632&interval=1d&events=history&includeAdjustedClose=true"
DJI = pd.read_csv(url)
DJI['date'] = pd.to_datetime(DJI['Date'])
DJI = DJI.drop(["Date"], axis = 1)

url = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/us.csv"
COVID = pd.read_csv(url)
COVID['date'] = pd.to_datetime(COVID['date'])
````

## Part A

Then, make a new array called `per_day_date` that gives the appropriate date for each case count. In particular, you will need to ensure that `per_day` and `per_day_date` have the same shape. 

````python
# your solution here

per_day = np.diff(COVID["cases"]) #use np.diff to get the number of new cases per day
per_day_date = COVID["date"][:-1] #make a copy of the dates except the last one so the shapes match
````
